# DARNet 消融实验分析

读取归档结果，对六种实验模式的 MAE、RMSE、RRSE 和 CORR 进行比较。

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'results' / 'all_modes_summary.csv').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
RESULT_PATH = PROJECT_ROOT / 'results' / 'all_modes_summary.csv'
results = pd.read_csv(RESULT_PATH)
results[['mode', 'MAE', 'RMSE', 'RRSE', 'CORR']].round(4)
```

```python
best = results.loc[results['RMSE'].idxmin()]
baseline_rmse = results.loc[results['mode'] == 'baseline', 'RMSE'].iloc[0]
improvement = (baseline_rmse - best['RMSE']) / baseline_rmse * 100
print(f"最佳模式: {best['mode']}")
print(f"最佳 RMSE: {best['RMSE']:.6f}")
print(f"相比 Baseline 降低: {improvement:.2f}%")
```

```python
ordered = results.sort_values('RMSE')
colors = ['#16a34a' if mode == ordered.iloc[0]['mode'] else '#3b82f6' for mode in ordered['mode']]
fig, ax = plt.subplots(figsize=(11, 5))
bars = ax.bar(ordered['mode'], ordered['RMSE'], color=colors)
ax.bar_label(bars, fmt='%.4f', padding=3)
ax.set(title='DARNet Ablation Study on ETTh1', xlabel='Mode', ylabel='RMSE (lower is better)')
ax.tick_params(axis='x', rotation=20)
ax.set_ylim(12.5, 14.0)
plt.tight_layout()
plt.show()
```

## 结论

在当前统一实验配置中，Patch 模式取得最低 RMSE。结果支持局部片段表示对短期油温预测的有效性；Attention、季节特征及其组合仍需要进一步调参与跨数据集验证。
